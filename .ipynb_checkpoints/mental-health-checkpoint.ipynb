```python
df = pd.read_csv("./dataset/survey.csv")
df.head()
```

```python
df.info()
```

```python
# Clean all null values.
df.state.fillna("", inplace=True)
df.work_interfere.fillna(df.work_interfere.mode()[0], inplace=True)
df.self_employed.fillna(df.self_employed.mode()[0], inplace=True)
df.comments.fillna("None", inplace=True)
print(df.isnull().sum())
```

```python
# Combine country and state
df["Country"].where(
    df["state"] == "", other=df["Country"] + ", " + df["state"], inplace=True
)
df.drop("state", axis=1, inplace=True)
df.head()
```

```python
df.Gender.value_counts()
```

```python
# Clean gender's and make them either Male, Female, Non-binary

df["Gender"].replace(
    [
        "A little about you",
        "Agender",
        "All",
        "Androgyne",
        "Enby",
        "non-binary",
        "Nah",
        "something kinda male?",
        "p",
        "ostensibly male, unsure what that really means",
        "Genderqueer",
        "queer/she/they",
        "Neuter",
        "Trans woman",
        "Trans-female",
        "queer",
        "fluid",
        "fluid",
        "male leaning androgynous",
        "Female (trans)",
        "Guy (-ish) ^_^",
    ],
    "Nonbinary",
    inplace=True,
)


df["Gender"].replace(
    [
        "Cis Female",
        "F",
        "Femake",
        "Female ",
        "Female (cis)",
        "Woman",
        "femail",
        "female",
        "woman",
        "cis-female/femme",
        "f",
    ],
    "Female",
    inplace=True,
)

df["Gender"].replace(
    [
        "Cis Male",
        "Cis Man",
        "M",
        "Mail",
        "Make",
        "Mal",
        "Male ",
        "Male (CIS)",
        "Male-ish",
        "Man",
        "m",
        "cis male",
        "maile",
        "male",
        "msle",
        "Malr",
    ],
    "Male",
    inplace=True,
)
df.Gender.value_counts()
```

```python
# Not normalized data
df.Age.describe()
```

```python
def filterByAge(age):
    indeces_list = df.index[df["Age"] == age].tolist()
    return df.loc[indeces_list]


filterByAge(5)
```

```python
# Remove any age outliers above the age 100.
df["Age"] = df["Age"].abs()
assert len(df[df["Age"] < 0]) == 0
indeces_list = df.index[df["Age"] > 100].tolist()
df = df.drop(indeces_list)
df.Age.describe()
```

```python
bins = [0, 18, 30, 40, 50, 60, 70, np.inf]
labels = ["0-17", "18-29", "30-39", "40-49", "50-59", "60-69", "70+"]
df["agerange"] = pd.cut(df.Age, bins, labels=labels, include_lowest=True)
df.agerange.value_counts().plot(
    kind="bar",
    xlabel="Age groups",
    ylabel="Number of people",
    title="Age group frequency",
)
plt.show()
```

```python
# Only United States
df_only_us = df.where(df["Country"].str.contains("United States")).dropna(axis=0)
df_only_us["Country"] = df["Country"].str.replace("United States, ", "")
df_only_us.Country.value_counts()[:10].plot(
    kind="bar",
    xlabel="States",
    ylabel="Number of US States",
    title="Top 10 US States frequency",
    figsize=(20, 5),
)
plt.xticks(rotation=360)
plt.show()
```

```python
# All Countires
df_global = df.copy()
df_global["Country"] = df.Country.str.replace(r"(^.*United States.*$)", "United States")
df_global.Country.value_counts()[:10].plot(
    kind="bar",
    xlabel="States",
    ylabel="Number of US States",
    title="US States frequency",
    figsize=(20, 5),
)

plt.show()
```

```python
df.drop("Timestamp", axis=1, inplace=True)
df.drop("agerange", axis=1, inplace=True)
```

```python
for column in list(df.columns.values):
    print(column, list(df[column].unique()))
```

```python
# One hot encoding
hot_encoding_features = [
    "self_employed",
    "family_history",
    "treatment",
    "work_interfere",
    "remote_work",
    "tech_company",
    "benefits",
    "care_options",
    "wellness_program",
    "seek_help",
    "anonymity",
    "leave",
    "mental_health_consequence",
    "phys_health_consequence",
    "coworkers",
    "supervisor",
    "mental_vs_physical",
    "mental_health_interview",
    "phys_health_interview",
    "obs_consequence",
]
for feature in hot_encoding_features:
    encoder = OneHotEncoder(handle_unknown="error")
    data_encoded = encoder.fit_transform(df[[feature]])
    for index, unique_value in enumerate(df[feature].unique()):
        unique_value = unique_value.lower().replace(" ", "_")
        col_name = feature + "_" + unique_value
        data_encoded_array = data_encoded.toarray()
        df[col_name] = data_encoded_array[:, index]
    df.drop(feature, axis=1, inplace=True)
df.head()
```
